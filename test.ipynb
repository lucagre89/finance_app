```python
import yfinance as yf
import streamlit as st
import pandas as pd
import numpy as np

assets = ["GOOGL", "AAPL", "MSFT"]

data = yf.download(assets, start="2025-05-31", end="2025-08-01", interval="1d")

print(data)
```

```python
df = pd.DataFrame(data[['Close', 'Volume']])
prices = data['Close']
prices.head()
```

```python
print(prices.iloc[:,0])
```

```python


weights = [0.3, 0.5, 0.2]  # Example weights for GOOGL, AAPL, MSFT

returns_df = prices.pct_change()

# Straightforward portfolio returns calculation
#returns_df['portfolio_check'] = returns_df.iloc[:,0] * 0.5 + returns_df.iloc[:,1] * 0.3 + returns_df.iloc[:,2] * 0.2


# Portfolio calculation with loop
portfolio_temp = []

for asset in assets:
    portfolio_temp.append(returns_df[asset] * weights[assets.index(asset)])


```

```python
print(portfolio_temp[0])

```

```python
len(pd.concat(portfolio_temp, axis=1).sum(axis=1))
```

```python
portfolio_returns = pd.concat(portfolio_temp, axis=1).sum(axis=1)
portfolio_returns.iloc[0] = np.nan
cumulative_returns = (1 + portfolio_returns).cumprod() - 1
```

```python
cumulative_returns
```

```python
# Assets covariance matrix
W = np.array(weights)
portfolio_std = (W.dot(returns_df.cov()).dot(W)) ** (1/2)
print(portfolio_std)
```
